```python
import matplotlib.pyplot as plt
import numpy as np
from udm_epic1 import SyntheticSampleGenerator, GeneratorConfig
from udm_epic1.augmentation.transforms import AugmentationPipeline, AugConfig

# Generate base sample
cfg = GeneratorConfig(height=256, width=256)
gen = SyntheticSampleGenerator(cfg, seed=1)
image, mask, _ = gen.generate('aug_demo')
img_f = (image.astype(float) / 65535.0).astype(np.float32)

aug = AugmentationPipeline(AugConfig(domain_shift_enabled=True, ring_artifact_prob=1.0))
img_aug, mask_aug = aug(img_f, mask)

fig, axes = plt.subplots(2, 2, figsize=(8, 8))
axes[0,0].imshow(img_f, cmap='gray'); axes[0,0].set_title('Original Image'); axes[0,0].axis('off')
axes[0,1].imshow(img_aug, cmap='gray'); axes[0,1].set_title('Augmented Image'); axes[0,1].axis('off')
axes[1,0].imshow(mask, cmap='hot'); axes[1,0].set_title('Original Mask'); axes[1,0].axis('off')
axes[1,1].imshow(mask_aug, cmap='hot'); axes[1,1].set_title('Augmented Mask'); axes[1,1].axis('off')
plt.suptitle('Domain-Shift Augmentation (Ring Artifact Simulation)')
plt.tight_layout()
plt.show()
```
