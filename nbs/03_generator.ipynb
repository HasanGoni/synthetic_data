```python
import matplotlib.pyplot as plt
import numpy as np
from udm_epic1 import SyntheticSampleGenerator, GeneratorConfig

cfg = GeneratorConfig(height=256, width=256)
gen = SyntheticSampleGenerator(cfg, seed=42)

fig, axes = plt.subplots(2, 4, figsize=(14, 7))
for i, (ax_img, ax_mask) in enumerate(zip(axes[0], axes[1])):
    image, mask, meta = gen.generate(f'sample_{i:04d}')
    img8 = (image.astype(float) / 65535.0 * 255).astype(np.uint8)
    ax_img.imshow(img8, cmap='gray', vmin=0, vmax=255)
    ax_img.set_title(f'{meta.n_voids} voids\narea={meta.total_void_area_fraction:.3f}')
    ax_img.axis('off')
    ax_mask.imshow(mask, cmap='hot')
    ax_mask.axis('off')
axes[0][0].set_ylabel('Image')
axes[1][0].set_ylabel('Mask')
plt.suptitle('Synthetic Samples — Beer-Lambert Physics')
plt.tight_layout()
plt.show()
```
