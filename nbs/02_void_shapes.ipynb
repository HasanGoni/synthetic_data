```python
import matplotlib.pyplot as plt
import numpy as np
from udm_epic1.generators.void_shapes import VoidShapeGenerator, VoidGeometry

rng = np.random.default_rng(42)
gen = VoidShapeGenerator(rng=rng)
shapes = ['ellipse', 'irregular_blob', 'elongated', 'cluster']

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for ax, shape in zip(axes, shapes):
    geom = VoidGeometry(cx=64, cy=64, area_fraction=0.06, shape=shape, contrast=0.2, edge_sigma=1.0)
    mask = gen.generate(128, 128, geom)
    ax.imshow(mask, cmap='hot')
    ax.set_title(shape)
    ax.axis('off')
plt.suptitle('Void Morphology Classes')
plt.tight_layout()
plt.show()
```
