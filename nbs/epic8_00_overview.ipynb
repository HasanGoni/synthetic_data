# Epic 8 -- Universal Model Support & Integration

Epic 8 is the **capstone** of the UDM synthetic data system.  It unifies all
previous epics (1-7) into a single cross-modality pipeline for generating,
exporting, and evaluating synthetic data across every supported inspection
modality.

## What you will learn

1. How the **Modality Registry** maps names to generators
2. The **Unified Pipeline** that orchestrates multi-modality generation
3. **Export** formats: COCO, YOLO, HuggingFace
4. **Cross-modality evaluation** and reporting

## Architecture

```
configs/epic8_universal.yaml
        |
        v
+-------------------+     +--------------------+
| UnifiedPipeline   |---->| ModalityRegistry   |
| (orchestrator)    |     | xray -> Epic 1     |
|                   |     | controlnet -> Epic 2|
|                   |     | cyclegan -> Epic 3  |
|                   |     | dann -> Epic 4      |
|                   |     | active -> Epic 5    |
|                   |     | aoi_wire -> Epic 6  |
|                   |     | chromasense -> Epic 7|
+-------------------+     +--------------------+
        |
        v
+-------------------+     +--------------------+
| Export            |     | Evaluation         |
| COCO / YOLO / HF |     | Cross-modality     |
+-------------------+     | Real vs Synthetic  |
                          +--------------------+
```

## Prerequisites

- Python 3.12, all UDM epic packages installed
- Install: `uv pip install -e ".[epic8]"`
