# Epic 8 -- Modality Registry

The `ModalityRegistry` is the central dispatch mechanism that maps modality
names (e.g. `xray`, `aoi_wire`) to their generator functions and config classes.

All previous epics are **pre-registered** at import time.

```python
from udm_epic8.registry.modality_registry import registry

# List all registered modalities
print("Available modalities:")
for name in registry.list_modalities():
    fn, cfg_cls = registry.get(name)
    print(f"  {name:15s} -> {fn.__name__:30s} config={cfg_cls.__name__}")
```

## Registering a Custom Modality

You can register your own modality generator:

```python
from pathlib import Path
from udm_epic8.registry.modality_registry import ModalityRegistry

# Create a fresh registry (or use the global one)
my_registry = ModalityRegistry()

def my_custom_generator(config: dict, n_samples: int, output_dir: str) -> Path:
    """Example custom generator."""
    out = Path(output_dir)
    out.mkdir(parents=True, exist_ok=True)
    print(f"Generating {n_samples} samples for custom modality -> {out}")
    return out

my_registry.register("custom_thermal", my_custom_generator, dict)
print(my_registry.list_modalities())
```

```python
# Generate using the registry
import tempfile

with tempfile.TemporaryDirectory() as tmp:
    result = my_registry.generate(
        "custom_thermal",
        config={"wavelength": "8-14um"},
        n_samples=10,
        output_dir=tmp,
    )
    print(f"Result: {result}")
```
