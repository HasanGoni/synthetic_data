# Epic 6 -- AOI Bond Wire & Surface Defect Synthesis

This notebook introduces the synthetic AOI image generation pipeline for
**bond wire defect detection** in semiconductor packages.

## Problem

Automatic Optical Inspection (AOI) systems detect defects in wire-bonded
semiconductor packages.  Common defects include:

| Defect | Description |
|--------|------------|
| **Bent wire** | Wire displaced laterally or sagging due to mechanical stress |
| **Broken wire** | Wire fractured, creating a gap in the connection |
| **Wire lift** | Bond pad detachment, one end of wire rises |

Training defect detectors requires large labelled datasets, which are expensive
to collect.  Epic 6 generates **fully synthetic** AOI images with pixel-level
ground truth masks.

## Architecture

```
Wire Geometry (Bezier)  -->  Defect Generator  -->  AOI Renderer  -->  Dataset
    BondWireProfile          bend/break/lift        background +       image + mask
    generate_wire_profile    apply_*_defect          wires + lighting   + metadata
```

## Modules

| Module | Description |
|--------|------------|
| `udm_epic6.models.wire_geometry` | `BondWireProfile`, Bezier-based wire generation |
| `udm_epic6.models.defect_generator` | `apply_bend_defect`, `apply_break_defect`, `apply_lift_defect` |
| `udm_epic6.rendering.aoi_renderer` | `render_aoi_image`, `render_background` |
| `udm_epic6.data.dataset` | `BondWireDataset`, `generate_bond_wire_dataset` |
| `udm_epic6.evaluation.metrics` | F1, IoU, wire detection rate, classification accuracy |
| `udm_epic6.cli_epic6` | CLI: generate, preview, evaluate, stats |
