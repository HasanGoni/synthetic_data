# Epic 9 — Synthetic Crack Generation: Overview

This epic implements three core capabilities for semiconductor crack defect synthesis:

## Capability 1: Synthetic Crack Generation from USM Images

Generate realistic crack patterns and overlay them onto Ultrasonic Microscopy (USM) images.
Cracks appear as bright lines in USM due to acoustic impedance mismatch.

**Crack types supported:**
- **Die crack** — radial cracks from corners/edges of silicon die
- **Substrate crack** — linear cracks following PCB stress lines
- **Mold crack** — network pattern in mold compound
- **Delamination crack** — interface-following cracks at layer boundaries

## Capability 2: Domain Transfer to RGB

Translate USM crack images to RGB domain appearance using:
- Physically-motivated colormaps (inferno, jet, etc.)
- Placeholder for learned CycleGAN-based transfer

## Capability 3: Mask-to-Image Generation

Given only a binary crack mask, generate a full synthetic image in either USM or RGB domain.
This is conditioned generation — the mask controls crack location and geometry.

```python
import numpy as np
import matplotlib.pyplot as plt

from udm_epic9.rendering.usm_renderer import generate_synthetic_usm_with_cracks
from udm_epic9.domain_transfer.usm_to_rgb import USMtoRGBTransfer, mask_to_image
```

```python
# Quick demo: generate USM with cracks, transfer to RGB, mask-to-image
rng = np.random.default_rng(42)

# Capability 1: USM crack generation
usm_img, mask, meta = generate_synthetic_usm_with_cracks(height=256, width=256, rng=rng)

# Capability 2: Domain transfer
transfer = USMtoRGBTransfer(method='colormap')
rgb_img = transfer.transfer_with_cracks(usm_img, mask)

# Capability 3: Mask-to-image
rgb_from_mask = mask_to_image(mask, target_domain='rgb')

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(usm_img, cmap='gray'); axes[0].set_title('USM Image')
axes[1].imshow(mask, cmap='gray'); axes[1].set_title('Crack Mask')
axes[2].imshow(rgb_img); axes[2].set_title('USM -> RGB Transfer')
axes[3].imshow(rgb_from_mask); axes[3].set_title('Mask -> RGB Image')
for ax in axes: ax.axis('off')
plt.tight_layout()
plt.show()
```
