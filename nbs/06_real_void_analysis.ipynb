## Purpose

Epic 1 compares synthetic voids to **real** annotations. This notebook mirrors `scripts/analysis/analyze_real_voids.py`: per-mask void counts and area fractions, saved as JSON for `scripts/validation/compare_distributions.py`.

**Convention:** mask PNGs are 8-bit; void pixels `> 127`.

````python
#| hide
from nbdev.showdoc import *
````

````python
import tempfile
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np

from udm_epic1.validation.mask_stats import scan_mask_directory, summarize_mask_records
````

````python
# Demo: two synthetic mask files (stand in for real data paths)
with tempfile.TemporaryDirectory() as tmp:
    d = Path(tmp)
    a = np.zeros((128, 128), dtype=np.uint8)
    a[30:70, 30:70] = 255
    b = np.zeros((128, 128), dtype=np.uint8)
    b[10:25, 10:25] = 255
    b[80:100, 60:90] = 255
    cv2.imwrite(str(d / "m0.png"), a)
    cv2.imwrite(str(d / "m1.png"), b)

    records, summary = scan_mask_directory(d, glob="*.png")
    print("summary:", summarize_mask_records(records))
    areas = [r["void_area_fraction"] for r in records]
    plt.figure(figsize=(5, 3))
    plt.bar(range(len(areas)), areas)
    plt.xlabel("mask index")
    plt.ylabel("void area fraction")
    plt.title("Per-mask void area (demo)")
    plt.tight_layout()
    plt.show()
````
