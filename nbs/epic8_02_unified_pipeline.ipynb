# Epic 8 -- Unified Pipeline

The `UnifiedPipeline` orchestrates synthetic data generation across multiple
modalities in a single run.  It reads a `UnifiedPipelineConfig` and dispatches
to the appropriate epic generator for each modality.

## Configuration

````python
from udm_epic8.pipeline.unified import UnifiedPipelineConfig

config = UnifiedPipelineConfig(
    modalities=["xray", "aoi"],
    per_modality_config={
        "xray": {"samples": 10},
        "aoi": {"samples": 5},
    },
    output_dir="/tmp/epic8_demo",
    total_samples=10,
    train_ratio=0.7,
    val_ratio=0.2,
    test_ratio=0.1,
)

print(f"Modalities: {config.modalities}")
print(f"X-ray samples: {config.samples_for('xray')}")
print(f"AOI samples:   {config.samples_for('aoi')}")
print(f"Split counts for 100 samples: {UnifiedPipelineConfig().compute_split_counts(100) if hasattr(UnifiedPipelineConfig, 'compute_split_counts') else 'N/A'}")
````

## Running the Pipeline

````python
from udm_epic8.pipeline.unified import UnifiedPipeline

pipeline = UnifiedPipeline(config)
manifest = pipeline.run()
print(f"Manifest: {manifest}")
````

````python
import json

with open(manifest) as f:
    data = json.load(f)

print(json.dumps(data, indent=2))
````

## Single Modality Run

You can also run a single modality directly:

````python
single_config = UnifiedPipelineConfig(
    modalities=["xray"],
    output_dir="/tmp/epic8_single",
    total_samples=5,
)

single_pipeline = UnifiedPipeline(single_config)
result = single_pipeline.run_modality("xray", n_samples=5)
print(f"Result: {result}")
````
