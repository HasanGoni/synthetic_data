# Epic 7 -- Defect Spectral Signatures

This notebook demonstrates how different defect types modify the spectral
reflectance of semiconductor materials.  Understanding these spectral
changes is key to training defect detection models on multi-spectral data.

## Defect types

1. **Delamination** -- air gap at material interface
2. **Contamination** -- foreign material deposit
3. **Oxidation** -- surface oxide layer

```python
import numpy as np
import matplotlib.pyplot as plt
from udm_epic7.spectral.wavelength_model import default_spectral_config, material_reflectance
from udm_epic7.spectral.defect_spectra import (
    delamination_spectrum,
    contamination_spectrum,
    oxidation_spectrum,
)
```

```python
cfg = default_spectral_config()
material = 'copper'

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# -- Delamination at varying severity --
ax = axes[0]
base = {wl: material_reflectance(material, wl, cfg) for wl in cfg.wavelengths}
ax.plot(list(base.keys()), list(base.values()), 'k-o', linewidth=2, label='Pristine')
for sev in [0.3, 0.6, 0.9]:
    spec = delamination_spectrum(material, sev, cfg)
    ax.plot(list(spec.keys()), list(spec.values()), 'o--', label=f'severity={sev}')
ax.set_title('Delamination', fontsize=13)
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Reflectance')
ax.legend()
ax.grid(True, alpha=0.3)

# -- Contamination --
ax = axes[1]
ax.plot(list(base.keys()), list(base.values()), 'k-o', linewidth=2, label='Pristine')
for cont in ['flux_residue', 'organic_film', 'dust']:
    spec = contamination_spectrum(cont, 0.6, material, cfg)
    ax.plot(list(spec.keys()), list(spec.values()), 'o--', label=cont)
ax.set_title('Contamination (conc=0.6)', fontsize=13)
ax.set_xlabel('Wavelength (nm)')
ax.legend()
ax.grid(True, alpha=0.3)

# -- Oxidation --
ax = axes[2]
ax.plot(list(base.keys()), list(base.values()), 'k-o', linewidth=2, label='Pristine')
for sev in [0.3, 0.6, 0.9]:
    spec = oxidation_spectrum(material, sev, cfg)
    ax.plot(list(spec.keys()), list(spec.values()), 'o--', label=f'level={sev}')
ax.set_title('Oxidation', fontsize=13)
ax.set_xlabel('Wavelength (nm)')
ax.legend()
ax.grid(True, alpha=0.3)

fig.suptitle(f'Defect Spectral Modifications -- {material}', fontsize=15, y=1.02)
fig.tight_layout()
plt.show()
```

**Key takeaways:**

- **Delamination** reduces reflectance with a wavelength-dependent pattern (more in blue)
- **Contamination** blends the base spectrum toward the contaminant's signature
- **Oxidation** uniformly reduces copper's IR reflectance, the strongest distinguishing feature
